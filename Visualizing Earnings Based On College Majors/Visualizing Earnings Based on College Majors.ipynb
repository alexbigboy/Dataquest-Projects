```python
ax1 = recent_grads[:10].plot.bar(x='Major', y='ShareWomen')
ax1.set_title("Share of Women in Top 10 Highest Paying Majors")

ax2 = fig.add_subplot(2,1,2)
ax2 = recent_grads[len(recent_grads)-10:len(recent_grads)].plot.bar(x='Major', y='ShareWomen')
ax2.set_title("Share of Women in Top 10 Lowest Paying Majors")

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,10))

numMen = recent_grads.pivot_table(index='Major_category',values='Men', aggfunc=np.sum)
numWomen = recent_grads.pivot_table(index='Major_category',values='Women', aggfunc=np.sum)

ind = np.arange(recent_grads['Major_category'].nunique())
width = .35

p1 = ax.bar(ind, numMen, width, color='r')
p2 = ax.bar(ind+width,numWomen, width, color='b')
ax.set_xticks(ind + width)
ax.set_xticklabels(recent_grads['Major_category'].unique(), rotation=90)
ax.legend(('Men','Women'))
```

```python
tbl = recent_grads.pivot_table(index='Major_category', values=['Men','Women'],aggfunc=np.sum)
tbl.plot.bar(x=tbl.index,y=['Men','Women'])
```
