# Star Wars Survey

```python
import pandas as pd
import numpy
star_wars = pd.read_csv("star_wars.csv", encoding="ISO-8859-1")
```

```python
star_wars.head(10)
```

```python
star_wars.columns
```

```python
star_wars = star_wars[star_wars["RespondentID"].notnull()]
```

```python
yes_no = {
    "Yes" : True,
    "No" : False
}
```

```python
star_wars['Have you seen any of the 6 films in the Star Wars franchise?'] = star_wars['Have you seen any of the 6 films in the Star Wars franchise?'].map(yes_no) 
```

```python
star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'] = star_wars['Do you consider yourself to be a fan of the Star Wars film franchise?'].map(yes_no)
```

```python
movie_names = {
    "Which of the following Star Wars films have you seen? Please select all that apply." : "seen_1",
    "Unnamed: 4" : "seen_2",
    "Unnamed: 5" : "seen_3",
    "Unnamed: 6" : "seen_4",
    "Unnamed: 7" : "seen_5",
    "Unnamed: 8" : "seen_6"
}
```

```python
star_wars[star_wars.columns[3:9]].values
```

```python
seen_true_false = {
    'Star Wars: Episode I  The Phantom Menace' : True,
    'Star Wars: Episode II  Attack of the Clones' : True,
    'Star Wars: Episode III  Revenge of the Sith' : True,
    'Star Wars: Episode IV  A New Hope' : True,
    'Star Wars: Episode V The Empire Strikes Back' : True,
    'Star Wars: Episode VI Return of the Jedi': True,
    numpy.NaN : False
}
```

```python
star_wars = star_wars.rename(columns=movie_names)
```

```python
for col in star_wars.columns[3:9]:
    star_wars[col] = star_wars[col].map(seen_true_false)
```

```python
star_wars.columns[9:15]
```

```python
rank_rename = {
    'Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.' : 'ranking_1',
    'Unnamed: 10' : 'ranking_2', 
    'Unnamed: 11' : 'ranking_3', 
    'Unnamed: 12' : 'ranking_4', 
    'Unnamed: 13' : 'ranking_5',
    'Unnamed: 14' : 'ranking_6'
}
```

```python
star_wars = star_wars.rename(columns=rank_rename)
```

```python
star_wars[star_wars.columns[9:15]] = star_wars[star_wars.columns[9:15]].astype(float)
```

```python
rank_means = star_wars[star_wars.columns[9:15]].mean()
```

```python
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

rank_means.plot.bar()
plt.show()
```

## Star Wars Ranks
The above bar graph shows the corresponding averange ranking for each of the Star Wars Movies. Based on averages the rank of the star wars movies is as follows :

1) Empire
2) Jedi
3) New Hope
4) Phantom
5) Clone Wars
6) Sith

## View counts
It appears that the original movies were seen by more respondents than the newer movies. This reinforces what we saw in the rankings, where the earlier movies seem to be more popular.

```python
males = star_wars[star_wars["Gender"] == "Male"]
females = star_wars[star_wars["Gender"] == "Female"]
```

```python
def star_wars_summary(data):
    rank_means = data[data.columns[9:15]].mean()
    print(rank_means)
    rank_means.plot.bar()
    plt.show()
    movie_seen = data[data.columns[3:9]].sum()
    print(movie_seen)
    movie_seen.plot.bar()
    plt.show()
    
def star_wars_summ_compare(data, data2):
    rank_means = data[data.columns[9:15]].mean()
    print(rank_means)
    rank_means.plot.bar()
    plt.show()
    rank_means = data2[data.columns[9:15]].mean()
    print(rank_means)
    rank_means.plot.bar()
    plt.show()
    
    movie_seen = data[data.columns[3:9]].sum()
    print(movie_seen)
    movie_seen.plot.bar()
    plt.show()
    movie_seen = data2[data.columns[3:9]].sum()
    print(movie_seen)
    movie_seen.plot.bar()
    plt.show()
```

```python
star_wars_summ_compare(males,females)
```

```python
for edu in star_wars['Education'].unique():
    print(edu)
    print('Total : ' + str(star_wars[star_wars['Education'] == edu]['Education'].count()))
    star_wars[star_wars['Education'] == edu][star_wars[star_wars['Education'] == edu].columns[9:15]].mean().plot.bar()
    plt.show()
```

```python
fig = plt.figure(figsize=(10,10))
for edu in star_wars['Education'].unique():
    if edu is not numpy.nan:
        ax1 = fig.add_subplot(3,3,int(numpy.where(star_wars['Education'].unique() == edu)[0])+1)
        ax1 = star_wars[star_wars['Education'] == edu][star_wars[star_wars['Education'] == edu].columns[9:15]].mean().plot.bar()
        ax1.set_title(edu)
        
plt.tight_layout(pad=0.4, w_pad=0.5, h_pad=1.0)
```
